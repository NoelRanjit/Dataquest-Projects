# Cleaning Data

We are trying to figure out the number of years employees have worked in theier companies and divide them into four categories:

New: Less than 3 years at a company

Experienced: 3-6 years at a company

Established: 7-10 years at a company

Veteran: 11 or more years at a company

```python
#checking the dtype of institute service
combined['institute_service'].dtypes
```

```python
# changing institute_service to string
combined['institute_service'] = combined['institute_service'].astype('str')
```

```python
# checking the year ranges
combined['institute_service'].value_counts().sort_values(ascending = False)
```

```python
# replacing vague numbers to a constant number
combined['institute_service'].replace('Less than 1 year', '0', inplace= True)
combined['institute_service'].replace('More than 20 years','21', inplace = True)
combined['institute_service'].value_counts()
```

```python
# splitting ranged numbers to a constant number
combined['institute_service'] = combined['institute_service'].str.split("-").str.get(0)
```

```python
# converting institute_service to float and checking to see if the replacements work correctly
combined['institute_service'] =combined['institute_service'].astype('float')
combined['institute_service'].value_counts(dropna= False)
```

```python
# categorizing on basis of career stage
def categorize(years_worked):
    if years_worked <3:
        return 'New'
    elif years_worked >=3 and years_worked <=6:
        return 'Experienced'
    elif years_worked >=7 and years_worked <=10:
        return 'Esablished'
    elif pd.isnull(years_worked):
        return np.nan
    else:
        return 'Veteran'

combined['service_cat'] = combined['institute_service'].apply(categorize)
```

```python
# Counting the experience range of the employees
combined['service_cat'].value_counts(dropna= False)
```

# Cleaning the Age columns

```python
# Checking the age values and their count 
combined['age'].value_counts()
```

We can see that all the age categories do not have the same formatting. One method of pairing them properly is to split them up and recategorize them.

```python
# Extracting the first part of the age range
combined['age'] = combined['age'].astype(str).str.extract('(\d+)').astype(float)
combined['age'].value_counts()
```

```python
# Reassigning age ranges
def age_ranges(age):
    if age >= 18 and age <=25:
        return '18 - 25'
    elif age >=26 and age <=30:
        return '26 - 30'
    elif age >=31 and age <=35:
        return '31 - 35'
    elif age >=36 and age <=40:
        return '36 - 40'
    elif age >=41 and age <=45:
        return '41 - 45'
    elif age >=46 and age <=50:
        return '46 - 50'
    elif age >=50 and age <=55:
        return '50 - 55'
    elif age >=56 and age <=60:
        return '56 - 60'
    elif age >=61:
        return '61 and above'
    elif pd.isnull(age):
        return np.nan
```

```python
# Using the funtion to sort ages and recounting ranges
combined['age_range'] = combined['age'].apply(age_ranges)
combined['age_range'].value_counts(dropna = False).sort_index()
```

```python
# Assigning values to pivot table
age_ranges = combined.pivot_table(index = 'age_range', values = 'dissatisfied')
age_ranges
```

```python
# Plotting disstisfaction based on age ranges
age_ranges.plot(kind = 'bar', legend = False)
```

The graph represents new and young workers to be in the most dissatisfied category. The dissatisfaction declines with more working experience. However, we see an uptick in the dissatisfaction amoung older employees. This could be attributed to older people also being new hires.
