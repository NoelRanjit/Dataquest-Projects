### Do students in more popular majors make more money?

```python
# Median incomes of students enrolled in popular majors
recent_grads.plot(x = "Total" , y = "Median" , kind = "scatter")
```

Students in more popular majors do not make more money. We can interpret from the plot that majors with less than 50,000 graduates make the most money with wages delclining the more popular a major becomes

### Are students majoring in subjects that are predominantly female paid more? 

```python
# Median incomes based on percentage of women in a subject
recent_grads.plot(x = "ShareWomen" , y = "Median" , kind = "scatter")
```

Students majored in subjects with majority females are shown to make less money. As the number of women increase in major, the wages made in the major decreases.

### Does more full time work mean more salary?

```python
# Link between full time employees and median salary
recent_grads.plot(x = "Full_time" , y = "Median" , kind = "scatter")
```

More full time workers does not mean more salary. As with popular majors, more full time workers mean less salary.

# Bar Plots

```python
men_majors = recent_grads.sort_values('ShareWomen', ascending = True)
men_majors[:10].plot.barh(x ='Major', y = 'Median')
men_majors[163:].plot.barh(x = 'Major', y = 'Median')
```

The major among men that makes the most wages is petroleum enginnering with early childhood education and communication disorders sciences and services making the least wages

```python
women_majors = recent_grads.sort_values('ShareWomen', ascending = False)
women_majors[:10].plot.barh(x = 'Major', y = 'Median', legend = False)
women_majors[163:].plot.barh(x = 'Major' , y = 'Median', legend = False)
```

Women majoring in Nursing and Health Medical Professions earned the most while they earned the least on Forestry and Transportaion Sciences/Technologies.

### Which category of majors have the most students?

```python
popular_majors = recent_grads.sort_values('Total', ascending = False)
popular_majors[:10].plot.barh(x = 'Major' , y = 'Total', legend = False)
popular_majors[163:].plot.barh(x = 'Major', y = 'Total', legend = False)
```

The most popular major is Psychology followed by Business Management and Administration. The least popular majors are Militart Technologies and Soil Science. Mathematics and Computer science is not the second least popular major since there are various mentions of mathematics and computer science related degrees in the list that fall under the same category and this does not represent the true number of people mejoring in the fields.

```python
unemployed_majors = recent_grads.sort_values('Unemployment_rate' , ascending = False)
unemployed_majors[:10].plot.barh(x = 'Major', y = 'Unemployment_rate', legend = False)
```

The majors most difficult to get jobs in are Nuclear Engineering followed by Public Administration
